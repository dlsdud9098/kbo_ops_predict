```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
import os
import seaborn as sns
from IPython.display import display
import pandas as pd


pd.options.display.max_columns = None
%matplotlib inline

# 한글 폰트
plt.rcParams['font.family'] = 'NanumGothic'
# 마이너스 깨짐 방지
matplotlib.rcParams['axes.unicode_minus'] = False

# 나눔 폰트 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'  # 나눔 고딕의 경로
font_prop = fm.FontProperties(fname=font_path).get_name()

# 한글 폰트 설정}
matplotlib.rcParams['font.family'] = font_prop

# 전역 팔레트 설정 (예: 'Set2' 팔레트 사용)
sns.set_palette('Set2')
```

```python
# 데이터 불러오기
regular_season = pd.read_csv('../datas/Regular_Season_Batter.csv')

print(list(regular_season['career']))
# 결측치 확인
regular_season.isna().sum()
```

```python
# 국적별 OPS 비교

'''
국적별로 비교해 보려고 한다.
한국인은 00초-00중 등 초 중 고 순으로 되어있지만 외국인은 그렇지 않다.
그것을 이용하여 한국인과 외국인을 나누려고 한다.
'''

# 국적별로 나눔
regular_season['country'] = regular_season['career'].apply(lambda x: '한국인' if '-' in x else x.split(' ')[0])

# 데이터 저장
regular_season.to_csv('../datas/Regular_Season_Batter.csv', encoding='utf-8-sig', index=False)
print(list(regular_season['country']))
```

```python
'''
국적별로 확인해 보았는데, 한국인에 비해 다른 국가에 소속된 선수들의 수가 너무 적다.
따라서 한국인, 외국인으로 나누려고 한다.
'''
regular_season['country'].value_counts()
```

```python
'''
네덜란드, 도미니카, 미국, 캐나다, 쿠바 -> 외국인 으로 수정
'''

regular_season['country'] = regular_season['career'].apply(lambda x: '한국인' if '-' in x else '외국인')

# 데이터 저장
regular_season.to_csv('../datas/Regular_Season_Batter.csv', encoding='utf-8-sig', index=False)
```

```python
regular_season['country'].value_counts()
```

```python
# 국적별 중앙값
median = regular_season.groupby(['country'])['OPS'].median().to_dict()

plt.figure(figsize=(20,9))
plt.subplot(1,2,1)

# 포지션별 OPS
ax = sns.boxplot(x='country', y='OPS', data=regular_season, showfliers = False)
plt.title('포지션별 OPS')
# 포지션별 중앙값
# median = regular_season.groupby('pos')['OPS'].median().to_dict()

# 박스 중앙에 글자 표시
for patch in ax.get_xticklabels():
    ax.text(patch.get_text(), median[patch.get_text()] + 0.03, f'Median: {median[patch.get_text()]:.1f}', 
            ha='center', va='center', fontsize=15, color='white')
    

plt.show()
```
