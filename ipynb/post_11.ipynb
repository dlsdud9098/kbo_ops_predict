```python
# 필요 없는 데이터 제거하고 열만 추출
df_cols = list(df.select_dtypes(include=['int64', 'float64']).drop(['batter_id', 'year', 'OPS', 'OBP', 'SLG'], axis=1).columns)

df_cols
```

```python
# 추출한 열과 연도, 선수 이름만 추출
df_temp = df[df_cols+['year', 'batter_name']].copy()

# 타수가 30 이상인 선수만 추출
df_temp = df_temp.loc[df_temp['AB'] >= 30]

df_temp.head()
```
